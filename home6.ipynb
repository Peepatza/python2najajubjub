```python
from abc import ABC, abstractmethod

class box(ABC):
    @abstractmethod
    def add(self, *args):
        pass
   
    @abstractmethod    
    def empty(self):
        pass

    @abstractmethod
    def count(self):
        pass

class item:
    def __init__(self, name, value):
        self.name = name
        self.value = value
    def __repr__(self):
        return f"({self.name}, {self.value})"

class listbox(box):
    def __init__(self):
        self.item = []
    def add(self, *item):
       self.item.extend(item)
    def empty(self):
        remove = self.item.copy()
        self.item.clear()
        return remove
    def count(self):
        return len(self.item)
    
class dictbox(box):
    def __init__(self):
        self.item = dict()
    def add(self, *item):
        for items in item:
            self.item[items] = items
    def empty(self):
        remove = list(self.item.values())
        self.item.clear()
        return remove
    def count(self):
        return len(self.item)

def repack_boxes(*args):
    all_item = []
    for box in args:
        all_item.extend(box.empty())
    
    total = len(all_item)
    number_boxes = len(args)
    minimum_item = total // number_boxes
    remainder = total % number_boxes

    index = 0
    box_index = 0
    for box in args:
        if box_index < remainder:
            item_to_add = minimum_item + 1
        else:
            item_to_add = minimum_item
        box.add(*all_item[index:index + item_to_add])
        index += item_to_add
        box_index += 1


box_a = listbox()
box_b = listbox()
box_c = dictbox()

box_a.add(*(item(chr(i+65), i+1) for i in range(20)))
box_b.add(*(item(chr(i+97), i+1) for i in range(9)))
box_c.add(*(item(i[0], i[1]) for i in (('ก', 67), ('ข', 68), ('ฃ', 69), ('ค', 70), ('ฅ', 71))))

repack_boxes(box_a, box_b, box_c)

print(box_a.count(), box_b.count(), box_c.count())
print(box_a.empty(), box_b.empty(), box_c.empty(), sep='\n')
        
```
