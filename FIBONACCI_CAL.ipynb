```python
import time

class timer:
    def __enter__(self):
        self.start = time.time()
        return self
    def __exit__(self, exc_type, exc_value, exc_tb):
        self.end = time.time()
        self.total = self.end - self.start
        print(f"total_time {self.total}")
        return False

def fibo(x):
    if x == 1 or x == 2:
        return 1
    else:
        return fibo(x - 1) + fibo(x - 2)
   
with timer() as t:
   fibo(35)

        
```
