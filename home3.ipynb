```python
class circle:
    pi =  3.14159
    def __init__(self, radius):
        self.radius = radius
    def get_area(self):
        return circle.pi * (self.radius ** 2)
    def get_circumference(self):
         return 2 * circle.pi * self.radius
c = circle(5)
print(c.get_area())
print(c.get_circumference())
```

```python
class rectangle:
    def __init__(self, lenght, width):
        self.lenght = lenght
        self.width = width
    def get_area(self):
        return self.lenght * self.width
    def get_perimeter(self):
        return 2 * (self.lenght * self.width)
rectangles = rectangle(10, 5)
print(rectangles.get_area())
print(rectangles.get_perimeter())
```

```python
class clock:
    def __init__(self, hour, minute):
        self.hour = hour
        self.minute = minute
    def add(self, other):
        new_minute = self.minute + other.minute
        if new_minute >= 60:
            extra_hour = new_minute // 60
            new_minute = new_minute % 60
        else:
            extra_hour = 0
        
        self.minute = new_minute

        new_hour = self.hour + other.hour + extra_hour

        if new_hour >= 24:
            new_hour = new_hour % 24
        
        self.hour = new_hour
    
    def display_time(self):
        print(self.hour, self.minute)
    
    def display_total_minute(self):
        print(self.hour * 60 + self.minute)
    
clock1 = clock(23, 30)
clock2 = clock(14, 20)
clock1.add(clock2)
(clock1.hour, clock1.minute)
(clock1.display_time())
(clock1.display_total_minute())   
```

```python
class bankaccount:
    def __init__(self, account_number, name, balance):
        self.account_number = account_number
        self.name = name
        self.balance = balance
    
    def put_money(self, money):
        self.balance = self.balance + money
    
    def withdraw(self, money):
        self.balance = self.balance - money
    
    def apply_bank_fees(self):
        self.balance = self.balance * 0.05
    
    def display(self):
        print("account number", self.account_number)
        print("Name", self.name)
        print("Balance", self.balance)

account = bankaccount(12343241, "InwPAT", 999)
account.put_money(700)
account.withdraw(200)
account.apply_bank_fees()
account.display()
```
