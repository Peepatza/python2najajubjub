```python
def n_time(n):
    def decorator(func):
        def wrapper():
            for _ in range(n):
                func()
        return wrapper
    return decorator

@n_time(3)
def print_my_name():
    print("Peepat")

print_my_name()
```

```python
import random
def once(func):
    run = False
    result = 0

    def wrapper():
        nonlocal run, result
        if not run:
            result = func()
            run = True
        return result
    return wrapper

@once
def random_number():
    return random.randint(1, 10)

print(random_number())
print(random_number())
print(random_number())

```
