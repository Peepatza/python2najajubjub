```python
def do_meaningless_things(argument):
    answer = 0
    array = [[1, 2, 3], [4.4, 5, "6"], ['list("78")', 9.0, 10/(argument-5)], [None]]
    for i in range(4):
        for j in range(argument):
            answer += array[i][j]
    return answer

try:
    do_meaningless_things(0) # argument can be anything
except ZeroDivisionError:
    print("You can't divide by zero")
except IndexError:
    print("Input made index ran out of the list")
except TypeError:
    print("Your input might be the wrong type")
else:
    print("everything is fine")

    
```
