```python
def do_meaningless_things(argument):
    answer = 0
    array = [[1, 2, 3], [4.4, 5, "6"], ['list("78")', 9.0, 10/(argument-5)], [None]]
    for i in range(4):
        for j in range(argument):
            answer += array[i][j]
    return answer

try:
    do_meaningless_things(0) # argument can be anything
except ZeroDivisionError:
    print("You can't divide by zero")
except IndexError:
    print("Input made index ran out of the list")
except TypeError:
    print("Your input might be the wrong type")
else:
    print("everything is fine")

    
```

```python
def get_country_name(phone_number:str) -> str:
    if not isinstance(phone_number, str):
        raise TypeError("Phone number is not a string")
    if phone_number[0] != "+":
        raise ValueError("Input is not a phone number")
    
    country_code_map = {"+1"  : "North American",
                        "+20" : "Egypt",
                        "+211": "South Sudan",
                        "+235": "Chad",
                        "+34" : "Spain",
                        "+66" : "Thailand",
                        "+91": "India"}
    
    if phone_number[0:2] in country_code_map.keys() and len(phone_number) == 12:
        return country_code_map[phone_number[0:2]]
    
    elif phone_number[0:3] in country_code_map.keys() and len(phone_number) == 13:
        return country_code_map[phone_number[0:3]]
    
    elif phone_number[0:4] in country_code_map.keys() and len(phone_number) == 14:
        return country_code_map[phone_number[0:4]]
    
    else:
        if 12 <= len(phone_number) <= 14:
            raise LookupError("Phone number doesn't contain any country code")
        raise ValueError("Invalid amount of phone digits")
    
#print(get_country_name(12345))
#print(get_country_name("66123456789"))
print(get_country_name("+77123456789"))
#print(get_country_name("+66123"))
```
